```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline

# Create random generator
rg = np.random.default_rng()
```

```python
# Number of arms
k = 10

# Mean reward of each arm
mu = rg.standard_normal(k)

# Which action is optimal
a_opt = np.argmax(mu)

# Time horizon
T = 10000

# Value of each action
Q = np.zeros(k)

# Number of times each action was taken
n = np.zeros(k)

# How greedy
epsilon = 0.5

log = {
    't': [],
    'r': [],
    'a': [],
    'a_is_opt': [],
}

for t in range(T):
    
    # Choose action
    if (rg.random() < epsilon):
        a = rg.integers(0, k)
    else:
        a = np.argmax(Q)
    
    # Simulate
    r = mu[a] + rg.standard_normal()
    
    # Update action value estimates
    n[a] += 1
    Q[a] += (1 / n[a]) * (r - Q[a])
    
    # Update log
    log['t'].append(t)
    log['r'].append(r)
    log['a'].append(a)
    log['a_is_opt'].append(1 if a == a_opt else 0)

plt.plot(log['t'], log['r'])
```

```python
def train(T):
    # Number of arms
    k = 10

    # Mean reward of each arm
    mu = rg.standard_normal(k)

    # Which action is optimal
    a_opt = np.argmax(mu)

    # Value of each action
    Q = np.zeros(k)

    # Number of times each action was taken
    n = np.zeros(k)

    # How greedy
    epsilon = 0.05

    log = {
        't': [],
        'r': [],
        'a': [],
        'a_is_opt': [],
    }

    for t in range(T):

        # Choose action
        if (rg.random() < epsilon):
            a = rg.integers(0, k)
        else:
            a = np.argmax(Q)

        # Simulate
        r = mu[a] + rg.standard_normal()

        # Update action value estimates
        n[a] += 1
        Q[a] += (1 / n[a]) * (r - Q[a])

        # Update log
        log['t'].append(t)
        log['r'].append(r)
        log['a'].append(a)
        log['a_is_opt'].append(1 if a == a_opt else 0)
    
    return log

# Time horizon
T = 1000

# Number of runs
n_runs = 100

r = np.zeros((n_runs, T))
a_is_opt = np.zeros((n_runs, T))

for i in range(n_runs):
    results = train(T)
    r[i, :] = results['r']
    a_is_opt[i, :] = results['a_is_opt']

r_avg = np.mean(r, axis=0)
a_is_opt_avg = np.mean(a_is_opt, axis=0)

t = range(T)
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
ax[0].plot(t, r_avg)
ax[1].plot(t, a_is_opt_avg)
# ax[0].set_y_label('average reward')
# ax[1].set_y_label('percent optimal action')
    
```
